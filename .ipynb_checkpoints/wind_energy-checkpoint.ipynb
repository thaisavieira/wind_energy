<center> <h1> Data Analysis of a Wind Turbine </h1></center>

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
```

```python
turbine = pd.read_csv('T1.csv')
turbine.columns = ['Date/Time', 'Power(kW)', 'Wind Speed(m/s)', 'Theoretical Curve(KWh)', 'Wind Direction (°)']
del turbine['Wind Direction (°)']
display(turbine)
```

## Graph of the real situation using active power (kW):

```python
sns.scatterplot(data=turbine, x='Wind Speed(m/s)', y=  'Power(kW)')
```

## Graph using theoretical data: 

```python
sns.scatterplot(data=turbine, x='Wind Speed(m/s)',y='Theoretical Curve(KWh)')
```

## Handling experimemental errors and uncertainties:

```python
active_power = turbine['Power(kW)'].tolist()
theoretical_power = turbine['Theoretical Curve(KWh)'].tolist()
# Handling the extremes, maximum,  and minimum power (P with 5% deviation)
max_power = []
min_power = []
within_limit = []

for power in theoretical_power:
    max_power.append(power*1.05)
    min_power.append(power*0.95)

# Case inside and outside the practical curve and also when it's zero
for p, power in enumerate(active_power):
    if power >= min_power[p] and power <= max_power[p]:
        within_limit.append('Inside')
    elif power == 0:
        within_limit.append('Zero')
    else:
        within_limit.append('Outside')

print(within_limit.count('Inside')/len(within_limit))
```

* With this we can conclude that approximately 37% is within the established limit/theoretical curve with error.

```python
# Adding list within_limit to dataframe:
turbine['Limit'] = within_limit
display(turbine)
```

## Plotting graph again (with limit range):

```python
colors = {'Inside':'blue', 'Outside':'red', 'Zero':'orange'}
sns.scatterplot(data=turbine, x='Wind Speed(m/s)',y='Power(kW)', hue='Limit', s=1, palette=colors)
```
